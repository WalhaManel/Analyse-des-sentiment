 Importation des biblioth√©ques

```python
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import names
from nltk.tokenize import word_tokenize
from nltk.tokenize import WordPunctTokenizer
```

Importation des fichiers et les fusionner dans une liste

```python
data1=open('amazon_cells_labelled.txt','r')
data2=open('imdb_labelled.txt','r')
data3=open('yelp_labelled.txt','r')
liste=[]

for d1,d2,d3 in zip(data1,data2,data3):
    liste.append(d1)
    liste.append(d2)
    liste.append(d3)

data1.close()
#afficher le nombres des lignes 
len(liste)
```

1- Tokenization et Normalisation

```python
import string 
import re
tk = WordPunctTokenizer()
def norm(liste):
    tk = WordPunctTokenizer()
    liste_r=[re.sub(r"[]"+ string.punctuation + r"]", " ",l) for l in liste]
    return liste_r
liste1=norm(liste)
data_words=[tk.tokenize(word.lower()) for word in liste1 ]
```

2- Stop-words

```python
from nltk.corpus import stopwords

def clean (data_words):
    mysp=stopwords.words('english')

   
    data_ready=[]
    for row in data_words:
        for w in row:
            words=[w for w in row if w not in mysp]
        data_ready.append(words)
    return data_ready
#data_ready
data_ready=clean(data_words)
```

3- Entrainement du mod√©le

```python
def word_feats(words):
    return dict([(word, True) for word in words])


train_set=[(word_feats(l[0:-1]),l[-1]) for l in data_ready ]
classifier = NaiveBayesClassifier.train(train_set)
```

4- Pr√©diction

```python
neg = 0
pos = 0
sentence = "I was shocked because no signs indicate cash only."
sentence = sentence.lower()

words = sentence.split(' ')
words=norm(words)
words=clean([words])
words
```

5- Classification des r√©sultats

```python
for word in words:
    classResult = classifier.classify( word_feats(word))
classResult
if classResult == '0':
    neg = neg + 1
if classResult == '1':
    pos = pos + 1
    
print('Positive: ' + str(float(pos)/len(words)))
print('Negative: ' + str(float(neg)/len(words)))
```
